# 1.3 Regular Expression (slide 17)

- mbox-short.txt can be downloaded from https://www.py4e.com/code3/
- rstrip( ) removes characters from the right based on the argument (a string specifying the set of characters to be removed).If the chars argument is not provided, all whitspaces on the right are removed from the string.


## 0) Initial approach
use search( ) function in regular expression

```python
import re
hand = open('mbox-short.txt')
for line in hand:
    line = line.rstrip()
    if re.search('From:', line):
        print(line)
```

## 1) Revised approach

 - “\S+@\S+” means one or more non-white space character, followed by “@”, followed by one or more non-white space character. 

```python
import re
hand = open('mbox-short.txt')
for line in hand:
    line = line.rstrip()
    x = re.findall('\S+@\S+', line)
    if len(x) > 0:
        print(x)
```

## 2) Further revised regular expression 

### Note: we have switched from “+” to “*”

- “*” means 0 or more non-white space characters. 
- “+” means 1 or more

If we use “+”, then the expression would have indicated we need at least two characters before and after the “@” character, which is different from the search pattern of the original regular expression (at least one character before and after “@”).

```python
import re
hand = open('mbox-short.txt')
for line in hand:
    line = line.rstrip()
    x = re.findall('[a-zA-Z0-9]\S*@\S*[a-zA-Z0-9]', line)
    if len(x) > 0:
        print(x)
```

3) To extract just the domain name of the email addresses, we can simply add parentheses to the regular expression, like this: '[a-zA-Z0-9]\S*@(\S*[a-zA- Z0-9])'

```python
import re
hand = open('mbox-short.txt')
for line in hand:
    line = line.rstrip()
    x = re.findall('[a-zA-Z0-9]\S*@(\S*[a-zA-Z0-9])', line)
    if len(x) > 0:
        print(x)
```

# 2.1.1 Retrieve a Web Page with Socket

Here is a simple program that makes use of socket to connect to a web server, utilising HTTP protocol to request for a document, and display what the server sends back.

```python
import socket
# socket.AF_INET: address family of the host is IPv4 -  host is a string 
#                 with a hostname like 'www.example.com' or an IPv4 address 
#                 like '10.1.2.3'. port is an integer.

# socket.SOCK_STREAM: refers to socket type - when using TCP, you’re reading 
#                     from a continuous stream of bytes

mysock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
mysock.connect(('data.pr4e.org', 80))

# \r = CR (Carriage Return) → Used as a new line character in Mac OS before X
# \n = LF (Line Feed) → Used as a new line character in Unix/Mac OS X
# \r\n = CR(Carriage Return) + LF(Line Feed) → Used as a new line character in Windows

cmd = 'GET http://data.pr4e.org/romeo.txt HTTP/1.0\r\n\r\n'.encode()
mysock.send(cmd)
while True:
    data = mysock.recv(512)
    if len(data) < 1:
        break
    print(data.decode())
mysock.close()
```

# 2.1.3 Parsing Web Pages using Regular Expressions

- Web scraping refers to writing a program to retrieve pages from the internet, then examine the data in those pages to extract relevant information.

```python
# Target to find if the home page of a website contains link to other web pages
import urllib.request
import re
url = 'http://www.bbc.com'
html = urllib.request.urlopen(url).read()
# “. +?” indicates that the match should be done in a “non-greedy” manner 
# i.e. to match the smallest possible matching string as long as it fulfils 
#      the pattern of one or more characters after the expression ‘http://‘
links = re.findall(b'href="(http://.+?)"', html)
for link in links:
    print(link.decode())
```

# 2.2.1 Pasing XML (slide 19)


```python
import xml.etree.ElementTree as ET
data = '''
 <person>
   <name>John</name>
   <phone type="intl">
      +1 81818181
    </phone>
    <email hide="yes"/>
 </person>
 '''
tree = ET.fromstring(data)
print('Name:', tree.find('name').text)
print('Attr:', tree.find('email').get('hide'))
```

## Looping through XML nodes
Often an XML has multiple nodes. In this case, we need to use a loop to process all the nodes. The following program loops through all the person nodes in the XML document:

```python
import xml.etree.ElementTree as ET
_data = '''
 <students>
  <persons>
   <person>
       <name>John</name>
       <phone type="intl">
      +1 81818181
        </phone>
        <email hide="yes"/>
    </person>
    <person>
          <name>Peter</name>
       <phone type="intl">
          +1 91919191
        </phone>
        <email hide="yes"/>
     </person>
  </persons>
 </students>
 '''
people = ET.fromstring(_data)
lst = people.findall('persons/person')
for person in lst:
    print (person.find('name').text)
    print (person.find('email').get('hide'))
    print (person.find('phone').get('type'))
```

# 2.2.2 Pasing Data from Web Services (slide 21)

```python
# python version 2.x : use urllib2 as we need to pass in headers information ('User-Agent': 'Mozilla/5.0' ) 
#                      like a normal web browser to access the data - data.gov.sg rejects url calls without 
#                      header information

# python version 3.x : use urllib.request()

import urllib, json

# The resources gives information on graduates’ salary by different universities. 
# The data provided by the web resources are all in JSON format.

url = '''
 https://data.gov.sg/api/action/datastore_search?resource_id=9326ca53-9153-4a9c-b93f-8ae032637b70
 '''

req = urllib.request.Request(url, headers={'User-Agent':'Mozilla/5.0'})
html = urllib.request.urlopen(req).read()

# The result is a dictionary file with nested arrays inside
data = json.loads(html)

print (len(data['result']['records'])) # Number of records
print (data['result']['records'][0].keys()) # which are the column names of each record from this resource
```

## 2.3.1 GET method

```python
import requests
URI = 'https://jsonplaceholder.typicode.com/posts/'
# return list of all students from the url (JSON list of all the posts in the database)
r = requests.get(URI)
print (r.text)
```

```python
# save the result into jason format
data = json.loads(r.text)
print ("number of records " + str(len(data)) )
```

## 2.3.2 POST method (create)

```python
data={'title': 'My title', 'body': 'My post texts', 'userId':
  1}
r = requests.post('https://jsonplaceholder.typicode.com/posts/', data=data)
print (r.text)
```

## 2.3.3 PUT method (update)

```python
# a JSON response so that we know the update operation is successful
data={'id':'10', 'title': 'foo', 'body': 'bar', 'userId': 1}
r = requests.put('https://jsonplaceholder.typicode.com/posts/10', data=data)
print (r.text)
```
